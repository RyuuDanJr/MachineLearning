Derivative and Integrals

```python
x= np.linspace(1,5,100000)
y=x**2
dydx= np.gradient(y,x)
y_int= np.cumsum(y)*(x[1]-x[0])
```

```python
plt.plot(x,y)
plt.plot(x, dydx)
plt.plot(x,y_int)
```

```python
theata= np.linspace(0,2*np.pi,10000)
r= 1 +3/4*np.sin(3*theata)
x= r*np.cos(theata)
y=r*np.sin(theata)
```

```python
plt.plot(x,y)
```

```python
A= 1/2*sum(r**2)*(theata[1]-theata[0])
A
```

```python
L= sum(np.sqrt(r**2+ np.gradient(r,theata)**2))*(theata[1]-theata[0])
L
```

```python
x= np.linspace(-2,2,1000)
y=np.linspace(-2,2,1000)
xv, yv = np.meshgrid(x,y)
f= np.exp(-(xv**2+yv**2))*np.sin(xv)
plt.contourf(xv,yv,f,levels= 100)
plt.colorbar()
```

```python
np.abs(f.ravel()).sum()*(np.diff(x)[0]* np.diff(y)[0])
```

```python
f[xv**2+ yv**2> .5**2]
```

```python
np.abs(f[xv**2+ yv**2> .5**2].ravel()).sum()*(np.diff(x)[0]*np.diff(y)[0])
```
