```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import tree
```

```python
file_path= '/content/Decision_Tree_ Dataset.csv'
cols= ['Initial_Payment', 'Last_Payment', 'Credit_Score','House_Number', 'Sum', 'Result']
df= pd.read_csv(file_path, names=cols)
```

```python
df= df.drop(0)
df= df.drop('Sum', axis= 1)
df.head()
```

```python

mod_df= df.convert_dtypes(float)
print(mod_df.describe())
print(mod_df.info())
```

Train, Test, Splitting data up

```python
## creating feature dataset and target dataset
## x = features y= target
X= mod_df.values[:, 0:4]
y= mod_df.values[:, -1]
## Splitting up the data now
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=.3, random_state=100)
clf_entropy= DecisionTreeClassifier(criterion='entropy', random_state=100,
max_depth=3, min_samples_leaf= 5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred= clf_entropy.predict(X_test)
y_pred
```

```python
print(f'Accuracy is {accuracy_score(y_pred, y_test)*100}')
```

```python
report= classification_report(y_pred, y_test)
print(report)
```

```python
report_matrix= confusion_matrix(y_pred,y_test)
print(report_matrix)
```

Random Forest Implantation

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
np.random.seed(0)
```

```python
iris= load_iris()
dataset= pd.DataFrame(iris.data, columns= iris.feature_names)
dataset.head()
```

```python
dataset['species']= pd.Categorical.from_codes(iris.target, iris.target_names)
dataset.head()
```

```python
dataset['is_train']= np.random.uniform(0,1,len(dataset)) <=.75
dataset.head()
```

```python
train, test= dataset[dataset['is_train']==True], dataset[dataset['is_train']==False]
print(f'Number of observations of Training Data {len(train)}')
print(f'Number of observations of Testing Data {len(test)}')
```

```python
features= dataset.columns[:4]
features
```

```python
y=pd.factorize(train['species'])[0]
y
```

```python
clf= RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[features],y)
```

```python
clf.predict(test[features])
```

```python
clf.predict_proba(test[features])[10:20]
```

```python
preds= iris.target_names[clf.predict(test[features])]
preds[0:25]
```

```python
test['species'].head()
```

```python
pd.crosstab(test['species'], preds, rownames= ['Actual Species'], colnames= ['Predicted Species'])
```
