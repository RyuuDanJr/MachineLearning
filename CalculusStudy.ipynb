```python
x,a,b,c = smp.symbols('x a b c', real= True)
f= x*smp.exp(a*smp.sin(x*b**3))*smp.log(c*smp.log(x**2))
f
```

```python
dfdx= smp.diff(f,x)
dfdx
```

```python
d4fdx4 = smp.diff(f,x,4)
d4fdx4
```
